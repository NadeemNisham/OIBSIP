# Import Libraries & Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('dataset/Unemployment in India.csv')
```

# Perform Data Analysis

```python
data.head()
```

```python
data.info()
```

## Overall Employment Trend

```python
data[' Date'] = pd.to_datetime(data[' Date'])
```

```python
data.info()
```

```python
yearly_unemployment = data.groupby(data[' Date'].dt.year)[' Estimated Unemployment Rate (%)'].mean()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(yearly_unemployment.index, yearly_unemployment.values)
plt.title('Overall Unemployment Trends in India')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.xticks(rotation=45)
plt.show()
```

## Unemployment by Region

```python
state_unemployment = data.groupby('Region')[' Estimated Unemployment Rate (%)'].mean().sort_values()
```

```python
plt.figure(figsize=(12, 6))
state_unemployment.plot(kind='bar')
plt.title('Unemployment Rate by State in India')
plt.xlabel('State')
plt.ylabel('Unemployment Rate')
plt.xticks(rotation=45)
plt.show()
```
